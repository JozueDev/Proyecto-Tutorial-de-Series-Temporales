```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("Aquifer_Auser.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.plot(x='Date', y=df.columns[1], figsize=(12,4), title='Volumen de agua')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Convertir fechas
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df.set_index('Date', inplace=True)
df = df.fillna(method='ffill')  
df = df.dropna()  
df.head()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error


target_col = df.columns[0]
serie = df[target_col]

# Entrenamiento/test split
train = serie[:-30]
test = serie[-30:]

# Modelo
model = ARIMA(train, order=(2, 1, 2))
model_fit = model.fit()

forecast = model_fit.forecast(steps=30)
rmse = np.sqrt(mean_squared_error(test, forecast))
print(f"RMSE: {rmse:.4f}")

# VisualizaciÃ³n
plt.figure(figsize=(10,5))
plt.plot(train.index, train, label="Train")
plt.plot(test.index, test, label="Test")
plt.plot(test.index, forecast, label="PredicciÃ³n")
plt.legend()
plt.grid()
plt.title("PredicciÃ³n de volumen de agua")
plt.show()
```

```python
import joblib
joblib.dump(model_fit, "modelo_arima_agua.pkl")
```
